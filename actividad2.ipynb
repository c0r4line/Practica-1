```python
import csv
import os
import pandas as pd
from datetime import datetime
import numpy as np
import time
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def abrir_csv():
    archivo_csv= 'eventos.csv'
    path = os.path.join(os.getcwd(), archivo_csv)
    data_set= pd.read_csv(path,encoding="utf-8")
    data_set['Estado'].fillna('sin datos', inplace=True)
    data_set['Palabra'].fillna('sin datos', inplace=True)
    
    return data_set
```
